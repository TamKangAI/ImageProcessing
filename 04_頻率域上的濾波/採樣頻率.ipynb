```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import zoom

# 
def load_image(image_path):
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    return image

# 
def downsample_image(image, sampling_rate):
    return zoom(image, sampling_rate)

# 
image_path = 'Lenna.jpg'  #  'your_image.jpg' 
original_image = load_image(image_path)

# 
sampling_rate_low = 0.2  # 
sampling_rate_high = 0.3  # 

# 
low_sampled_image = downsample_image(original_image, sampling_rate_low)
high_sampled_image = downsample_image(original_image, sampling_rate_high)

# 
plt.figure(figsize=(12, 6))

# 
plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(original_image, cmap='gray')
plt.axis('off')

# 
plt.subplot(1, 3, 2)
plt.title(f"Low Sampling Rate ({sampling_rate_low})")
plt.imshow(low_sampled_image, cmap='gray')
plt.axis('off')

# 
plt.subplot(1, 3, 3)
plt.title(f"High Sampling Rate ({sampling_rate_high})")
plt.imshow(high_sampled_image, cmap='gray')
plt.axis('off')

plt.show()
```
